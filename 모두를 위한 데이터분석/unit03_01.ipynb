```python
import csv
max_temp=-999
max_date=''
f=open('seoul.csv','r',encoding='cp949')
data=csv.reader(f,delimiter=',')
header=next(data)
for row in data:
    if row[-1]=="":
        row[-1]=-999
    row[-1]=float(row[-1])
    if row[-1]>max_temp:
        max_temp=row[-1]
        max_date=row[0]
print('기상 관측 이래 서울의 최고 기온이 가장 높았던 날은',max_date+'로,', max_temp,'도 였습니다.')
f.close()
```

```python
import matplotlib.pyplot as plt
plt.title('title label(legend) color') # 그래프 이름
plt.plot( [10,20,30,40], color='pink',label='x1' ) #색은 color=생략해도 지정가능
plt.plot([12, 43, 25, 15], 'r',label='x2') #label값 정해주고 legend함수 호출해야 범례표시
plt.legend(loc=10) #loc 범례위치 지정
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('linestyle')
#빨간색 dashed 그래프
#plt.plot([10, 20, 30, 40],linestyle='--',label='dashed')
plt.plot([10, 20, 30, 40],'r--',label='dashed')
#dotted 그래프
plt.plot([40, 30, 20, 10],ls=':',label='dotted')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('marker')
plt.plot([10,20,30,40],'r.',label='circle') #빨간 원형
plt.plot([40, 30, 20, 10],'g^--',label='triangle up') #그린 삼각형 dashed
plt.legend()
plt.show()
```
