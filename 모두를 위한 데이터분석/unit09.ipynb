```python
# 항아리 그래프 만들기 using barh
import csv
import matplotlib.pyplot as plt

f=open('gender.csv')
data=csv.reader(f)

m=[]
f=[]
loc=input('검색을 원하는 지역 입력 : ')

for row in data:
    if loc in row[0]:
        for i in range(101):
            m.append(-int(row[i+3].replace(",",""))) # male 3~103  # -부호로 반대편에 보이게함
            f.append(int(row[-(i+1)].replace(",",""))) # female -101~-1
        break # 한 지역만 검색할 때. 중복 append 방지

f.reverse() # -1(100세이상)~ -101(0세) 뒤집기

plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title(loc+' 성별인구분포')
plt.barh(range(101), m, label='male')
plt.barh(range(101), f, label='female')
plt.legend()
plt.show()
```

```python
# pie
import matplotlib.pyplot as plt

size=[2441,2312,1031,1233]
label=['A','B','AB','O']
color=['#cccccc','deeppink','pink','skyblue']

plt.axis('equal') # 완전 circle
plt.pie(size, labels=label, autopct='%.1f%%', colors=color, explode=(0,0,0.1,0)) 
#autopct=퍼센트 표시 labels=이름표시 colors=color지정 explode=튀어나오기정도 지정
plt.legend() #범례 표시
plt.show()
```

```python
#제주도 성별 인구 pie
import csv


f = open('gender.csv')
data = csv.reader(f)

size = [0,0]
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data :
    if name in row[0] :
        for i in range(101) :
            size[0] += int(row[i+3].replace(',',''))    # 3 ~ 103번까지 데이터 더하기
            size[1] += int(row[-(i+1)].replace(',',''))  # 106 ~ 206번까지 데이터 더하기
        break #제주 들어가는 첫번째만 보고 for문 out

plt.rc('font',family='Malgun Gothic')
plt.axis('equal')
plt.pie(size, labels=['m','f'], colors=['pink','skyblue'],
        autopct='%.1f%%', startangle=90)
plt.title(name+' gender ratio')
plt.show()
#startangle : pie chart 시작점 3시로부터 반시계방향으로 angle만큼 설정
```
