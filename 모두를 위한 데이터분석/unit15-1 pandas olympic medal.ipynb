```python
#하계올림픽 데이터 추출
import pandas as pd
# read_html : 웹 페이지에서 테이블 형태의 데이터를 추출하는 함수
df= pd.read_html('https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table',
                header=0, index_col=0) 
#header < row index,*번째 row가 column header / index_col < col index,0번째 col이 col index

summer=df[1].iloc[:,:5] #iloc[row 범위,col 범위]

#col 이름 변경
summer.columns=['경기수','금','은','동','계']

# '금' 기준 내림차순(descending) / inplace true: 자체 내에서 정렬, false: 정렬된 dataframe 반환
summer.sort_values('금', ascending=False, inplace=True)

summer.to_excel('하계올림픽메달.xlsx')

summer
```
