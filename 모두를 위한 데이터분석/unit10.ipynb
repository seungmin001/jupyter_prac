```python
#성별, 연령 비교
import csv
import matplotlib.pyplot as plt

f=open('gender.csv')
data=csv.reader(f)

m=[]
f=[]
loc=input('검색하고픈 지역 입력 : ')

for row in data:
    if loc in row[0]:
        for i in range(101):
            m.append(int(row[i+3].replace(",",""))) #3~103
            f.append(int(row[-(i+1)].replace(",",""))) #-101~-1
        break
f.reverse()

plt.plot(m, label='male')
plt.plot(f, label='female')
plt.legend()
plt.show()
```

```python
# 숫자 차이로 막대그래프로 비교
import csv
import matplotlib.pyplot as plt

f=open('gender.csv')
data=csv.reader(f)

result=[]
loc=input('검색하고픈 지역 입력 : ')

for row in data:
    if loc in row[0]:
        for i in range(101):
            result.append(int(row[i+3].replace(",",""))-int(row[i+3+103].replace(",","")))
        break

plt.bar(range(101),result, label='m-f sub')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

x = []
y = []
size = []

for i in range(100) :
    #랜덤 생성
    x.append(random.randint(50,100))
    y.append(random.randint(50,100))
    size.append(random.randint(10,100))
    
plt.scatter(x, y, s=size, c=size, cmap='jet', alpha=0.7) #크기에 따라 색 부여
#alpha 0transparent~1opaque
plt.colorbar()
plt.show()
```

```python
#제주도 연령대별 성비 scatter로 표현
import csv
import math

f = open('gender.csv')
data = csv.reader(f)

m = []
f = []
size=[]
name = input('궁금한 동네를 입력해주세요 : ')

for row in data :
    if name in row[0] :
        for i in range(3,104) :
            mv=int(row[i].replace(",",""))
            fv=int(row[i+103].replace(",",""))
            m.append(mv)
            f.append(fv)
            size.append(math.sqrt(mv+fv))
        break
        
plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')
plt.title(name+' gender population graph')

plt.scatter(m, f, c=range(101), alpha=0.5, cmap='jet', s=size)
plt.colorbar()
plt.plot(range(max(m)),range(max(m)),'g') #y=x 추세선

plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.show()
```
