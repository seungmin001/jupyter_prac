```python
import csv
import matplotlib.pyplot as plt

f=open('age.csv')
data=csv.reader(f)
header=next(data)
#print(header)

quest=input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해 주세요 : ')
result=[]

for row in data:
    if quest in row[0]: # 0: 지역명, 1,2: 전체 인구수, 3~:나이별 인구수
        #print(len(row[3:])) # 101개 맞는지 확인
        for i in row[3:]:
            result.append(int(i))

print((result))
print(plt.style.available) # 사용가능한 plot style
plt.style.use('bmh')
plt.rc('font',family='Malgun Gothic') #한글 깨져서 폰트 설정
plt.title(quest+" 지역의 인구 구조")
plt.plot(result)
plt.show()
```
