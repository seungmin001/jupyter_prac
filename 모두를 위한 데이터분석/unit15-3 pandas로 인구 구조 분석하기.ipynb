```python
# 궁금한 지역 입력받고 해당 지역 인구구조 저장
import pandas as pd
import numpy as np

df= pd.read_csv('age.csv', encoding='cp949', index_col=0)
df= df.div(df['2020년08월_계_총인구수'], axis=0) 
del df['2020년08월_계_총인구수'],df['2020년08월_계_연령구간인구수']

name=input('원하는 지역의 이름 입력 : ')
# df.index.str.contains(). index 중 원하는 string이 포함된 row를 찾아냄
# return boolean series
a= df.index.str.contains(name) 
print(a)
# indexing 관련 궁금증
# https://stackoverflow.com/questions/48409128/what-is-the-difference-between-using-loc-and-using-just-square-brackets-to-filte
# loc iloc [row] [row,col]
df2=df[a] # condition
print(df2)

import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
df2.T.plot() # dataframe에도 plot기능 있음(df의 데이터로 이루어진 matplotlib의 plot을 반환)
plt.show()
```

```python
# 궁금한 지역의 인구 구조와 가장 비슷한 인구 구조를 가진 지역 시각화하기
import pandas as pd
import numpy as np

df= pd.read_csv('age.csv', encoding='cp949', index_col=0)
df= df.div(df['2020년08월_계_총인구수'], axis=0) 
del df['2020년08월_계_총인구수'],df['2020년08월_계_연령구간인구수']

name= input('원하는 지역의 이름 입력 : ')
a= df.index.str.contains(name)
df2= df[a] # 여기서 원하는 row 만 있는 'DataFrame' 획득

# 그 'row의 값'이 필요하기 때문에 dataframe에서 iloc으로 값 얻음
x= df.sub(df2.iloc[0], axis=1) # x: 인구구조 차이 dataframe
y= np.power(x,2) #차이를 알기 위해 abs또는 power필요
z= y.sum(axis=1) #각 row 마다 모든 col의 값 합 구하기

# 인구구조 차이 가장 작은 지역 5곳
# sort_values() : pandas function. axis default=0(index)
# dataframe.index는 df의 index들을 반환하고 slicing[:5]으로 그 중 5곳의 index를 얻어냄
i= z.sort_values().index[:5] 


import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
df.loc[i].T.plot() #위에서 얻은 index이용
plt.show()

#전체과정 한 줄로
df.loc[np.power(df.sub(df2.iloc[0], axis=1), 2).sum(axis=1).sort_values().index[:5]].T.plot()
```
