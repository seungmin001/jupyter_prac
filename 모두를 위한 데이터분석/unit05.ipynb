```python
# 생일 기온변화 그래프 작성하기(1983 이후)
import csv
import matplotlib.pyplot as plt
f=open('seoul.csv','r',encoding='cp949')

data=csv.reader(f,delimiter=',')
header=next(data)

high=[] #최고기온 데이터 저장할 리스트
low=[]

for row in data:
    if row[-1]!="" and row[-2]!='':  #최고기온 데이터 값 존재할때
        date=row[0].split('-') #중복사용 날짜 값 변수로 처리
        if int(date[0]) >=1983:  # after1983
            if date[1]=='08' and date[2]=='19':  #if 8.19
                high.append(float(row[-1])) #최고기온
                low.append(float(row[-2]))  #최저기온

                
                
plt.rc('font',family='Malgun Gothic') #한글 깨져서 폰트 설정
plt.rcParams['axes.unicode_minus']=False #마이너스 기호 깨짐 방지
plt.title('생일 기온변화 그래프')        
plt.plot(high,'hotpink',marker='.',label='최고기온')
plt.plot(low,'skyblue',marker='^',label='최저기온')
plt.legend()
plt.show()

f.close()
```

```python
# 평균 기온이 상승하는 그래프 표현

import csv
import matplotlib.pyplot as plt
f=open('seoul.csv','r',encoding='cp949')

data=csv.reader(f,delimiter=',')
header=next(data)

avg=[]
year=[]
for row in data:
    if row[-1]!="" and row[-2]!='':  #최고기온 존재 데이터만 취급
        date=row[0].split('-') # 연 월 일 list
        if date[1]=='08' and date[2]=='19':
            year.append(int(date[0]))
            high=float(row[-1])
            low=float(row[-2])
            avg.append((high+low)/2)

plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False

plt.title('생일 평균 기온 관찰')
plt.plot(year,avg,marker='.',label='평균 기온')
plt.legend()
plt.show()

f.close()
```
