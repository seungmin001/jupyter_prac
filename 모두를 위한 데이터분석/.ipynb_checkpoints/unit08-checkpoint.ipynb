```python
import csv
import matplotlib.pyplot as plt

f=open('age.csv')
data=csv.reader(f)

result=[]
for row in data:
    if '신도림' in row[0]:
        for i in row[3:]:
            result.append(int(i))

#plt.bar(range(101),result) #101개의 나이 구간
plt.barh(range(101),result) #수평 막대그래프. range가 y축
plt.show()
```

```python
# 항아리 그래프 만들기 using barh
import csv
import matplotlib.pyplot as plt

f=open('gender.csv')
data=csv.reader(f)
m=[]
f=[]
loc=input('검색을 원하는 지역 입력 : ')

for row in data:
    if loc in row[0]:
        for i in range(101):
            m.append(-int(row[i+3])) # male 3~103  # -부호로 반대편에 보이게함
            f.append(int(row[-(i+1)])) # female -101~-1

f.reverse() # -1(100세이상)~ -101(0세) 뒤집기

plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title(loc+' 성별인구분포')
plt.barh(range(101), m, label='male')
plt.barh(range(101), f, label='female')
plt.legend()
plt.show()
```
