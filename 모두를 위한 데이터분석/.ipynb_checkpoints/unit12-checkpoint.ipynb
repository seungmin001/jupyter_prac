```python
#출근시간 승차인원
import csv
import matplotlib.pyplot as plt

f=open('subwaytime.csv')
data=csv.reader(f)
next(data)
next(data)

result=[]

for row in data:
    #int변환
    row[4:]=map(int,row[4:]) #map(일괄적용할 함수이름, 함수를 적용할 데이터)
    #10열(7시data)
    result.append(row[10])
    
#print(len(result))
result.sort()
plt.style.use('ggplot')
plt.bar(range(len(result)),result)
plt.show()
# 한 역의 승차인원이 다른곳보다 월등하게 많다.
# 85%정도인 500개 역의 오전7시 승차인원이 5만명이 안된다.
```

```python
# 7~9 승차인원 합
import csv
import matplotlib.pyplot as plt

f=open('subwaytime.csv')
data=csv.reader(f)
next(data)
next(data)

result=[]
max_v=0
max_station=""

for row in data:
    #int변환
    row[4:]=map(int,row[4:]) #map(일괄적용할 함수이름, 함수를 적용할 데이터)
    #10열(7시data)
    result.append(sum(row[10:15:2])) #10~14까지 2간격 합 추가
    #최대값 역 찾기
    if max_v<sum(row[10:15:2]):
        max_v=sum(row[10:15:2])
        max_station=row[1]+" "+row[3]
    
print(max_station,max_v)

result.sort()
plt.style.use('ggplot')
plt.bar(range(len(result)),result)
plt.show()
```

```python
#7~9 가장 많은 인원이 하차하는 역
import csv

f=open('subwaytime.csv')
data=csv.reader(f)
next(data)
next(data)

max_v=0
max_station=""

for row in data:
    #int변환
    row[4:]=map(int,row[4:])
    #최대값 역 찾기
    a=sum(row[11:16:2]) #하차인원 추출 후 합
    if max_v<a:
        max_v=a
        max_station=row[1]+" "+row[3]
    
print(max_station,max_v)
```

```python
#특정 시간에 사람들이 가장 많이 승차하는 역
import csv

f=open('subwaytime.csv')
data=csv.reader(f)
next(data)
next(data)

max_v=0
max_station=""
t=int(input("시간 입력 : "))
i=4+(t-4)*2 # 시간에 맞는 index

for row in data:
    #int변환
    row[4:]=map(int,row[4:])
    #최대값 역 찾기
    a=row[i]
    if max_v<a:
        max_v=a
        max_station=row[1]+" "+row[3]
    
print(max_station,max_v)
```

```python
#시간대별로 사람들이 가장 많이 타고 내리는 역
import csv
import matplotlib.pyplot as plt

f=open('subwaytime.csv')
data=csv.reader(f)
next(data)
next(data)

max_v=[0]*24 #시간대별 max value
max_station=['']*24 #시간대별 station of max value

for row in data:
    #int변환
    row[4:]=map(int,row[4:])
    #최대값 역 찾기
    for i in range(24):
        a=row[4+i*2] # 시간대별 승차 인원
        if max_v[i]<a:
            max_v[i]=a
            max_station[i]=row[1]+" "+row[3]+'('+str(i+4)+'시)'
    
plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')
plt.bar(range(24),max_v)
#x축 이름 바꾸기 (set lable location, text label, properties)
plt.xticks(range(24), max_station, rotation=90) 
plt.show()
```

```python
#시간대별 최다 하차인원
import csv
import matplotlib.pyplot as plt

f=open('subwaytime.csv')
data=csv.reader(f)
next(data)
next(data)

max_v=[0]*24 #시간대별 max value
max_station=['']*24 #시간대별 station of max value

for row in data:
    #int변환
    row[4:]=map(int,row[4:])
    #최대값 역 찾기
    for i in range(24):
        b=row[4+i*2+1] # 시간대별 하차 인원
        if max_v[i]<b:
            max_v[i]=b
            max_station[i]=row[1]+" "+row[3]+'('+str(i+4)+'시)'
    
plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')
plt.bar(range(24),max_v,color='skyblue')
plt.xticks(range(24), max_station, rotation=90) #x축 이름 바꾸기
plt.show()
```

```python
#시간대별 모든 승하차 인원 
import csv
import matplotlib.pyplot as plt

f=open('subwaytime.csv')
data=csv.reader(f)
next(data)
next(data)

s_in=[0]*24 #시간대별 승차인원
s_out=[0]*24 #시간대별 하차인원

for row in data:
    #int변환
    row[4:]=map(int,row[4:])
    #시간대별 승하차 더하기
    for i in range(24):
        s_in[i]+=row[4+i*2]
        s_out[i]+=row[5+i*2]

plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in,label="승차 인원")
plt.plot(s_out,label="하차 인원")
plt.xticks(range(24),range(4,28))
plt.legend()
plt.show()
```
