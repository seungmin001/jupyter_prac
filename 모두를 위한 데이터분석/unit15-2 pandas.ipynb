```python
import pandas as pd
import numpy as np
index = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.rand(8,3), index=index, columns=list('ABC'))


#df column access = 대괄호에 column이름 넣기
print('col access')
print(df['A']) # Series형태 index=기본 인덱스, 1차원 배열형태

#mask기능 (return Series of boolean values. 원하는 데이터 골라내기)
print('\nmask')
print(df['B'] > 0.4)

#B열의 값이 0.4보다 큰 값이라는 조건이 True인 데이터(row select)들로만 이루어진 데이터 프레임
#Series of boolean values can be used to filter the DataFrame by putting it in between the selection brackets []. Only rows for which the value is True will be selected.
#source : https://pandas.pydata.org/docs/getting_started/intro_tutorials/03_subset_data.html
df2=df[df['B'] > 0.4]
print('\nmask adapted table')
print(df2)

#행과 열 뒤집기 Transpose
print('\ntranspose table')
print(df.T)
```

```python
# 2 dimensional array, DataFrame operation
# row 방향 연산 axis = 0
# col 방향 연산 axis = 1
import pandas as pd
import numpy as np
index = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.rand(8,3), index=index, columns=list('ABC'))

# A열의 값을 B열의 값으로 나눈 값을 D열에 저장
df['D']= df['A']/df['B']

# 행 우선(row major ordering,가로로) 계산 값을 E열에 저장 / 'columns'방향
df['E'] = np.sum(df, axis=1)

print(df.head())

# 열 우선(column major ordering, 세로로) 계산 / 'index'방향
# A열 값 기준으로 전체 데이터에 대한 뺄셈 (sub)
# 모든 열의 N번째 행 값에서 'A'열의 N번째 행 값을 뺄셈
df= df.sub(df['A'], axis=0)

print(df.head()) # 앞 5개

# 데이터 전체를 C열 값으로 나누기 (div)
df=df.div(df['C'], axis=0)

# save in csv
df.to_csv('test.csv')
print(df.head())
```
