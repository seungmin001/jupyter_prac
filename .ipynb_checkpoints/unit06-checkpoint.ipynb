```python
#주사위 시뮬레이션
import random
import matplotlib.pyplot as plt

dice=[]
for i in range(5):
    dice.append(random.randint(1,6))
print(dice) #1~6 중 하나

plt.hist(dice,bins=6) # bins 가로축 구간 개수 설정
plt.show()
```
