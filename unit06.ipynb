```python
#1월 8월 데이터 시각화
import matplotlib.pyplot as plt
import csv

f = open('seoul.csv')
data = csv.reader(f)
next(data)

jan = []
aug=[]

for row in data :
    month=row[0].split('-')[1]
    if row[-1] != '' : #최고기온 존재
        if month=='08': #8월
            aug.append(float(row[-1]))
        if month=='01': #1월
            jan.append(float(row[-1]))
            
        
plt.hist(aug,bins=100, color='r',label='aug')
plt.hist(jan,bins=100,color='b',label='jan')
plt.legend()
plt.show()
```

```python
# 정확한 위치 값 numpy로 알아내기
import numpy as np
import matplotlib.pyplot as plt
import random

result=[]
for i in range(13):
    result.append(random.randint(1,1000))

result=np.array(result)
print("1/4: "+str(np.percentile(result,25)))
print("2/4: "+str(np.percentile(result,50)))
print("3/4: "+str(np.percentile(result,75)))
```

```python
# jan & aug temp in boxplot
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv')
data=csv.reader(f)
next(data)
jan=[]
aug=[]

for row in data:
    month=row[0].split('-')[1]
    if row[-1]!='':
        if month=='08':
            aug.append(float(row[-1]))
        if month=='01':
            jan.append(float(row[-1]))
            
plt.boxplot([aug,jan])
plt.show()

# 동그라미 > outlier 다른 수치에 비해 너무 크거나 작은 값
```

```python
# monthly data
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv')
data=csv.reader(f)
next(data)
month=[[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data:
    m=row[0].split('-')[1]
    if row[-1]!='':
        month[int(m)-1].append(float(row[-1]))
            
plt.boxplot(month)
plt.show()
```

```python
# find august highest temp day
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv')
data=csv.reader(f)
next(data)
day=[]
for i in range(31):
    day.append([]) # 일별 데이터 추가할 

for row in data:
    m=row[0].split('-')[1]
    d=row[0].split('-')[2]
    if row[-1]!='':
        if m=='08':
            day[int(d)-1].append(float(row[-1]))

plt.style.use('ggplot') # 그래프 스타일 지정
plt.figure(figsize=(10,5), dpi=300) # 그래프 크기 수정
plt.boxplot(day, showfliers=False) # 아웃라이어(동그라미) 값 생략

plt.show()
```

```python
# 1년 중 일교차가 가장 큰 달은 몇 월일까요?
# 각 달으로 나누고 다시 각 일마다 일교차 계산 후 달 위치에 삽입
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv')
data=csv.reader(f)
next(data)
diff=[]
for i in range(12):
    diff.append([])

for row in data:
    month=int(row[0].split('-')[1])
    if row[-1]!="" and row[-2]!="":
        diff[month-1].append(float(row[-1])-float(row[-2]))

max=0
max_idx=-1
for i in range(12):
    temp=sum(diff[i])/len(diff[i])
    if temp>max:
        max=temp
        max_idx=i+1
        
print(str(max_idx)+"월의 일교차가 가장 큼")
```
